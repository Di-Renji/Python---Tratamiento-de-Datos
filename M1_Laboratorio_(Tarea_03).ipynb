# ***AGRUPAR Y GRAFICAR***

1) generos con las canciones mÃ¡s largas de duraciÃ³n mÃ¡s largos  (agrupar por gÃ©nero y sacar mÃ¡ximo de canciÃ³n)

```python
filtro_ayg_1 = data.loc[:, ('title', 'top genre', 'dur')]
agrupar_ayg_1 = filtro_ayg_1.groupby("top genre").max()  
agrupar_ayg_1.head(5)
```

```python
nuevo = agrupar_ayg_1.head(3)
genero = "top genre"
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(nuevo.title, nuevo.dur)
plt.show()
```

2) AÃ±os con su promedio de bailable (agrupar por aÃ±o y sacar el promedio de dnce)

```python
filtro_ayg_2 = data.loc[:, ('year', 'dnce')]
agrupar_ayg_2 = filtro_ayg_2.groupby("year").mean()  
agrupar_ayg_2
```

3) aÃ±os con su promedio de ruido (dB)

```python
filtro_ayg_3 = data.loc[:, ('year', 'dB')]
agrupar_ayg_3 = filtro_ayg_3.groupby("year").mean()  
agrupar_ayg_3
```

# ***HEAD / TAIL***

1) Sacar los 3 aÃ±os con mÃ¡s mÃºsica mÃ¡s ruidosa (promedio dB)

```python
filtromasruidosa = data.loc[:,("year", "dB")]  
agruparaniomasruidosa = filtromasruidosa.groupby("year").mean()  
ordenarmasruidosa = agruparaniomasruidosa.sort_values(by="dB", ascending=False)
ordenarmasruidosa.head(3) 
```

2) Sacar los 3 aÃ±os con mÃ¡s mÃºsica mÃ¡s bailable (promedio dnce)

```python
filtromasbailable = data.loc[:,("year", "dnce")]     # filtro para reportar solo el "year" y el "dnce"
agruparaniomasbailable = filtromasbailable.groupby("year").mean()   # agrupa los aÃ±os y saca el promedio de "dnce" por cada aÃ±o
ordenarmasbailable = agruparaniomasbailable.sort_values(by="dnce", ascending=False) # Ordnea por mas bailable
ordenarmasbailable.head(3) # Reportar los 3 aÃ±os mas bailables
```

3) Sacar los 3 artistas con mÃºsica mÃ¡s bailable (promedio dnce)

```python
filtromasbailable2 = data.loc[:,("artist", "dnce")]  
agruparaniomasbailable2 =filtromasbailable2.groupby("artist").mean()  
ordenarmasbailable2 = agruparaniomasbailable2.sort_values(by="artist", ascending=False)
ordenarmasbailable2.head(3) 
```

4) Sacar los 3 artistas con mÃºsica mÃ¡s popular  (promedio popular)

```python
filtromaspopular = data.loc[:,("artist", "pop")]  
agruparartistamaspop = filtromaspopular.groupby("artist").mean()  
ordenarmaspopular = agruparartistamaspop.sort_values(by="artist", ascending=False)
ordenarmaspopular.head(3) 
```
